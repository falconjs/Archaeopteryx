## Definitions

<p>A *hypothesis* is a statement about a population parameter or a sample statistic subject to verification.  In practical terms is defines the research question to be answered by the collection of data and the statistical analysis thereof.  Two hypotheses should be stated at the commencement of any research, namely the *null hypothesis* and the *alternate hypothesis*.  The null hypothesis is the statement of no departure from an expected result and the alternate hypothesis  is a statement to be accepted when statistical analysis shows a probability so extreme as to force rejection of the null hypothesis, by assumption of the fact the the finding could not have occurred by mere chance.  In healthcare research, this (the alternate hypothesis) would be the statement that is accepted if the p-value is less than what was originally decided on as a cut-off for statistical significance.  In most cases, this would be 0.05 (the alpha-value).  Wow, I hope you got most of that!</p>
<p>Following the statement of the research hypotheses, they should undergo a process of hypotheses testing.  This refers to a statistical procedure that uses sample evidence and probability theory to determine whether a statement about the value of a population parameter should be rejected or not.</p>
<p>Let's try to clarify the definitions stated above, considering an example.  A statement is made that the mean value of a new laboratory test used in the management of a specific disease is *18*.  A researcher decides to randomly select a group of patients and subject them to this test and calculates and average for her sample subjects of *21*.  It could logically be argued that the original statement was false and should be rejected.  To be sure, this would be incorrect, as statistical analysis should be performed first and more specifically, calculations should be performed to look at the probability of having found a value as extreme as *21*.</p>
<p>In essence the value *18* will be taken as a mark on the x-axis of a bell-shaped curve, which is formed from all the possible means that could have been calculated should all the possible random samples have been analyzed as has been explained in the lecture on the Central Limit Theorem.  The value *21* will fall somewhere on the x-axis.  See the schematci representation below.  Since *21* is larger than *18*, it would naturally fall to the right of *18* and if the area under the curve from *21* towards the positive side could be calculated, this will be the probability of finding a value as extreme as *21*.</p>

<p>To make more sense of this example, requires the introduction of the *alpha-value*.  In healthcare research, this would most often be 0.05 and signifies firstly, the risk the researcher is willing to take to falsely reject the null hypothesis, known as a *type 1 error*.  In more practical terms, it is the cut-off for clinical significance.  If the *p-value* (probability) of the statistical analysis is less than the alpha-value, the null hypothesis is rejected (statistically significant difference).</p>  <p>Visually, this *0.05* represents an area under the curve and can be placed in three possible locations depending on how the null hypothesis was stated.  The first possibility is the most common and is shown in the image above.  Here the null hypothesis, as per our example, would be that the mean for a sample should equal *18*.  This means the alternate hypothesis states that the mean is not *18*.  It is clear from this that the alternate hypothesis refers to values both above and below *18*.  In essence, the alpha value of *0.05* must be split in two, with *0.025* on either side. A mean value calculated by the researcher for her sample that falls in these shaded areas would result in that value’s area (probability) being less than *0.05*.</p>
<p>The second and third possibilities come from other ways of stating the null hypothesis.  The researcher might have stated as a null hypothesis, that the value that she is researching is larger than *18*.  She would now have to visually entertain, what is referred to as a one-tail right-sided test and all of the shaded area would fall to the right.  This would be akin to the image below.  If her initial alternate hypothesis was that a score must be less than *18*, it would be a one-tail left test, with all of the *0.05* area falling in the left side.</p>
<p>This is why the original statement of the research hypothesis is so important.  It makes a difference as to where the shaded areas fall and directly impacts whether the researcher’s calculated mean value will fall within the chosen alpha-value or not.  For this reason it is vitally important to state the hypotheses before initiating any research so as to remain an ethical researcher.</p>

## Some important points

<p>The result summary should always reflect the initial hypotheses statements as I have shown in the example above.  It is also important to note that the null hypothesis cannot be accepted.  Only two possible outcomes are allowed: either fail to reject the null hypothesis or reject the null hypothesis (thereby accepting the alternate hypothesis).  Furthermore, although in common medical use, the terms *statistically significant or not* are actually frowned upon.  It is more correct to suggest a likelihood of having found a value as extreme as the research has.  Alas, these terms are imbedded in the medical literature and we accept them as such.</p>
<p>You ,may now start to realize that a *p-value* is not some magical genie, granting our wish for a clear-cut answer from our research!  It depends on so many things, not least of which is the important notion that in order to use it, we have to make assumptions.  The most important one is probably the fact that we assume the null hypothesis to be true.  Only if it is true can we infer that our specific outcome was so rare to occur that it represent a significant finding.  If the null hypothesis was never true, our house of cards actually comes tumbling down.  Fortunately, or unfortunately, we never talk about this, though.  It is the deep dark secret of statistics.</p>
