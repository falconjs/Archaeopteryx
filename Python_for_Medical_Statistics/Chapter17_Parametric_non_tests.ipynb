# Parametric vs non-parametric statistical tests

## Introduction

<p>Before, I mentioned that a histogram of the mean of all possible means will be normally distributed.  Actually things are a bit more complicated than that.  No surprise!  In the grander scheme of things, there are two types of statistical tests, called parametric and non-parametric tests.  These two groups exist, because using the appropriate type of test for an appropriate setting allows for more accurate anlaysis and a more proper representation of what we are trying to achieve (infering statistics on population parameters).</p>

## The QQ plot

<p>A parametric test is used when the sample data is taken from a normal distribution and a non-parametric test if it is not.  Wow!  We only have our sample data.  How can we know this?  Well, there's the probability plot from the *scipy.stats* library.  It is more commonly known as the QQ-plot.  It plots every data point against its quartile on a Cartesian plane.  It then gives you a visual representation of how correlated these are.  If they are not, the data is not from a normal distribution and you have to use a non-parametric test.</p>

<p>First, I'll generate 40 random values from the normal distribution, and constrain it to a mean of 100 and a standand deviation of 20.</p>

```python
rv1_values = stats.norm.rvs(loc = 100, scale = 20, size = 40)
```

<p>Now let's do the QQ plot</p>
<p>The *stats.probplot()* function takes three arguments.  The first is the acutal data values, the second teels the software to compare our values to normal distribution.  If I wanted to see if my valuyes fit some other distribution I could have used something other than 'norm' and lastly we plot.</p>
<p>The plot will state a R<sup>2</sup> value.  A value of 1.0 means an absolute fit and 0.0 means no fit (i.e. realy, really not from a normal distribution.</p>

```python
plt.figure(figsize = (10,8))
stats.probplot(rv1_values, dist="norm", plot=plt) # 
plt.show();
```

<p>Now I'll generate 40 values from anotehr type of continuous random distribution called the exponential distribution and see how it compares.</p>

```python
rv2_values = stats.expon.rvs(size = 40)
```

```python
stats.probplot(rv2_values, dist="norm", plot=plt);
```

<p>Now we only get an *0.9482* fit.  It's really difficult to say what the exact cut-off is.  Clearly an *R<sup>2</sup>* value of *0.9482* indicates that the sample was not taken from a normal distribution.</p>
